# Feature Engineering

```python
# import required libraries
import numpy as np
import pandas as pd
```

```python
# read dataset
df = pd.read_csv("falcon9_dataset_transformed.csv")
df.head()
```

```python
# select the required features
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial', 'Class']]
features.head()
```

```python
# create dummy variables
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])
features_one_hot.head()
```

```python
# cast all numeric columns to float
features_one_hot.astype(float)
```

```python
# export as a csv file
features_one_hot.to_csv('falcon9_dataset_encoded.csv', index=False)
```
